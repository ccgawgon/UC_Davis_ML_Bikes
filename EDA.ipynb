```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv('SeoulBikeData.csv', encoding='ISO-8859-1')

numeric_df = df.select_dtypes(include='number')


correlation_matrix = numeric_df.corr()


sns.heatmap(correlation_matrix, annot=True, cmap='plasma', fmt=".2f", square=True) #  cbar_kws={'shrink': .8}
plt.title('Correlation Heatmap of Seoul Bike Data')
plt.show()
```

```python
plt.hexbin(df['Temperature(°C)'], df['Rented Bike Count'], gridsize=30, cmap='Purples')
plt.colorbar(label='Count in bin')
plt.title('Temperature vs. Rented Bike Count (2D Histogram)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rented Bike Count')
plt.show()

plt.hist(df['Rented Bike Count'], bins=30, alpha=0.7, color='purple', edgecolor='black')
plt.title('Histogram of Rented Bike Count')
plt.xlabel('Rented Bike Count')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
df['Log Rented Bike Count'] = np.log1p(df['Rented Bike Count']) 
plt.subplot(1, 2, 2)
plt.hist(df['Log Rented Bike Count'], bins=30, alpha=0.7, color='pink', edgecolor='black')
plt.title('Log Transformed Rented Bike Count')
plt.xlabel('Log Rented Bike Count')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
df['Sqrt Rented Bike Count'] = np.sqrt(df['Rented Bike Count'])

plt.subplot(1, 2, 2)
plt.hist(df['Sqrt Rented Bike Count'], bins=30, alpha=0.7, color='yellow', edgecolor='black')
plt.title('Square Root Transformed Rented Bike Count')
plt.xlabel('Square Root Rented Bike Count')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Humidity(%)'], df['Dew point temperature(°C)'], color='#cf5757', alpha=0.5)
plt.title('Humidity vs. Dew Point Temperature')
plt.xlabel('Humidity (%)')
plt.ylabel('Dew Point Temperature (°C)')
plt.grid(True)
plt.show()


plt.hist2d(df['Humidity(%)'], df['Dew point temperature(°C)'], bins=30, cmap='Reds')
plt.colorbar(label='Counts')
plt.title('2D Histogram of Humidity vs. Dew Point Temperature')
plt.xlabel('Humidity (%)')
plt.ylabel('Dew Point Temperature (°C)')
plt.grid(True)
plt.show()
```
